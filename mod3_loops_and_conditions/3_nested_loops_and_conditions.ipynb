## FizzBuzz Problem

Write a short program that prints each number from 1 to 100 except:

- For each multiple of 3, print "Fizz" instead of the number

- For each multiple of 5, print "Buzz" instead of the number

- For numbers which are multiples of both 3 and 5, print "FizzBuzz" instead of the number

### Q.1 Please outline in words the steps that should be executed for each iteration of the loop

Start by using modulus to define each number that has a remainder of 0 when divided by 3 as "Fizz". Then, do the same but when divided by 5, and define it as "Buzz". Finally, use "and" to make a nunber with a remainder of 0 when divided by 3 and 5 be defined as "FizzBuzz".

### Q.2 Please implement the solution described in Q.1 with comments

```python
for num in range(1,100):
    # define a number divisible by 3 as "Fizz" bc this means its a multiple.
    if num%3 == 0:
        print(f"Fizz")
    # define a number divisible by 5 as "Buzz" bc this means its a multiple.
    if num%5 == 0:
        print(f"Buzz")
    # define a number divisible by both 5 and 3 as "FizzBuzz" bc this means its a multiple.
    if num%5 and num%3 == 0:
        print(f"FizzBuzz")
    else:
        print(num)
```

### Q.3 Please complete the commented lines below with descriptions of the code as indicated. 

```python
# Create list to store prime numbers
prime_numbers_list = []

# This for loop iterates over n between the range of 2 and 11.
for n in range(2,11):
    # Set is_prime as True
    is_prime = True
    # This for loop iterates over k in the range of 2 and n.
    for k in range(2, n):
        # This condition checks if n is divisible by k.
        if n % k == 0:
            # If n % k == 0, then print the expression that shows n is divisible by k.
            print(f"{n} = {k} * {n // k}")
            is_prime = False
            break
    # This condition checks if is_prime is True
    if is_prime:
        # If n is prime, then print as output 
        # and append to list of prime numbers
        print(f"{n} is a prime number")
        prime_numbers_list.append(n)
        
print(f"\nList of prime numbers in range: {prime_numbers_list}")
```
